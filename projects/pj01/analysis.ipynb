First I will read in the csv data and convert it into a column-orientated format.

```python
from data_utils import read_csv_rows, columnar

data_rows: list[dict[str, str]] = read_csv_rows('../../data/survey.csv')
columns: dict[str, list[str, str]] = columnar(data_rows)
columns
```

Now that the data is read in I will select the year columns, and list the first 5 rows using head.

```python
from data_utils import head, select
condensed: dict[str, list[str]] = select(columns, ["year"])
distilled: dict[str, list[str]] = head(condensed, 5)
distilled
```

From here we can tell that the functions work and give us the years of each student. Next we will analyze the number of people in each year using counts.

```python
from data_utils import count
year_counts: dict[str, int] = count(condensed["year"])
year_counts
```

As we can see years 23, and 24 have the highest counts followed by 25, then 22, and finally 21. This shows that sophmores and juniors show the highest numbers, while my original hypothesis predicted freshmen and sophmores to be the highest concentration, very interesting. Now for 2 cool graphs to fulfill the above and beyond requirement...

```python
from matplotlib import pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(12, 5), sharey=True)
fig.suptitle("Students in COMP110 sorted by year.")


axes[0].set_title("Bar chart of class year in COMP110")
axes[0].bar(year_counts.keys(), year_counts.values())
axes[0].tick_params(axis='x', labelrotation = 45)

axes[1].set_title("Scatter chart of class year in COMP110")
axes[1].scatter(year_counts.keys(), year_counts.values())
```

Aditionally if we take the average of the bottom two classes and the top three, considering that year 21 are the same as seniors, we will find...

```python
from data_utils import avg
years_21_22_23: list[int] = [10, 74, 114]
years_24_25: list[int] = [282, 93]
print(avg(years_21_22_23))
print(avg(years_24_25))
```
